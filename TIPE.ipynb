#I Neurones et Réseaux de neurones



###Définitions:

> **Un Neurone** est une représentatation mathématique et informatiques d'un neurone biologique. Il contient en general plusieurs entrées et une seule sortie. <br>
Mathématiquement, un neurone est une fonction à plusieurs variables et à valeurs réelles.

>**Un réseau de Neurones** est une association de neurones pour accomplir des tâches arbitrairement complexes.

>**Le Cout $C$** d'un reseau mesure la difference entre la valeur obtenue et la valeur souhaitée: 
$C=\frac{1}{2}(a-y)^2$ où a est la valeur du dernier neurone (dans le cas d'une derniere couche a 1 neurone) et y la valeur souhaitée pour ce neurone

>**$\hat{X}$** est une matrice contenant des valeurs pour les neurones de la première couche et **$\hat{Y}$** les valeurs attendues en sortie.

>**$α$** Est la vitesse d'apprentissage, un paramètre à choisir

##Notations:


> *   $w_{i,j}^\ell$: le poids reliant le neurone $i$ de la couche $l$ au neurone $j$ de la couche $l+1$
*   $z_{i}^\ell$: la valeur de la somme ponderée avant passage dans la fonction d'activation
*   $b_{i}^\ell$: la valeur du biais $i$ de la couche $l$
*   $σ:$ Fonction d'activation à valeurs dans un intervalle réduit, à dérivée peu couteuse en calcul. Elle permet de rendre les calculs du réseau non-linéaires
*   $a_{i}^\ell$: la valeur du neurone $i$ de la couche $l$. Elle vaut $\sigma(z_{i}^\ell)$
*   \begin{align}
A_\ell =
\begin{pmatrix} a_1^\ell & a_2^\ell & ... & a_n^\ell
\end{pmatrix}
\end{align}
*   \begin{align}
Z_\ell =
\begin{pmatrix} z_1^\ell & z_2^\ell & ... & z_n^\ell
\end{pmatrix}
\end{align}
*   \begin{align}
B_\ell =
\begin{pmatrix} b_1^\ell & b_2^\ell & ... & b_n^\ell
\end{pmatrix}
\end{align}
*    \begin{align}
W_\ell =
\begin{pmatrix} w_{1, 1}^\ell & w_{1,2}^\ell & ... & w_{1,k}^\ell \\ 
w_{2,1}^\ell & w_{2,2}^\ell & ... & w_{2,k}^\ell \\
... & ... & ... & ... \\
w_{n,1}^\ell & w_{n,2}^\ell & ... & w_{n,k}^\ell
\end{pmatrix}
\end{align}
Où $n$ est le nombre de neurones dans la couche $\ell$ et $k$ dans la couche $\ell+1$
*
\begin{align}
∇_w^{\ell}=
\begin{pmatrix} \frac{∂}{\partial w_{1, 1}^\ell} & \frac{∂}{\partial w_{1,2}^\ell} & ... & \frac{∂}{\partial w_{1,k}^\ell} \\ 
\frac{∂}{\partial w_{2,1}^\ell} & \frac{∂}{\partial w_{2,2}^\ell} & ... & \frac{∂}{\partial w_{2,k}^\ell} \\
... & ... & ... & ... \\
\frac{∂}{\partial w_{n,1}^\ell} & \frac{∂}{\partial w_{n,2}^\ell} & ... & \frac{∂}{\partial w_{n,k}^\ell}
\end{pmatrix}
\end{align}
*\begin{align}
∇_b^{\ell}=
\begin{pmatrix} \frac{∂}{\partial b_1^\ell} & \frac{∂}{\partial b_2^\ell} & ... & \frac{∂}{\partial b_n^\ell}
\end{pmatrix}
\end{align}
*\begin{align}
∇_z^{\ell}=
\begin{pmatrix} \frac{∂}{\partial z_1^\ell} & \frac{∂}{\partial z_2^\ell} & ... & \frac{∂}{\partial z_n^\ell}
\end{pmatrix}
\end{align}
*\begin{align}
∇_a^{\ell}=
\begin{pmatrix} \frac{∂}{\partial a_1^\ell} & \frac{∂}{\partial a_2^\ell} & ... & \frac{∂}{\partial a_n^\ell}
\end{pmatrix}
\end{align}
*\begin{align}δ^\ell=∇_z^\ell C
\end{align}
* Produit Hadamard:
\begin{align}
\begin{pmatrix} a_{1,1} & a_{1,2} \\
a_{2,1} & a_{2,2}
\end{pmatrix}
\circ
\begin{pmatrix} b_{1,1} & b_{1,2} \\
b_{2,1} & b_{2,2}
\end{pmatrix}=
\begin{pmatrix} a_{1,1} \cdot b_{1,1} & a_{1,2} \cdot b_{1,2} \\
a_{2,1} \cdot b_{2,1} & a_{2,2} \cdot b_{2,2}
\end{pmatrix}
\end{align}

 


#Formules:


> Propagation avant: 
\begin{align} Z_{\ell+1}= A_\ell \cdot W_\ell+B_{\ell+1}
\end{align}
\begin{align} A_{\ell+1}= \sigma(Z_{\ell+1})
\end{align}

> Propagation arrière:

>\begin{align}
\frac{\partial C}{\partial z_j^\ell}=\frac{\partial C}{\partial a_j^\ell} \cdot \frac{\partial a_j^\ell}{\partial z_j^\ell}=\frac{\partial C}{\partial a_j^\ell} \cdot \sigma'(z_j^\ell)
\end{align}
\begin{align}
δ^\ell=∇_a^\ell C \circ \sigma'(Z_\ell)
\end{align}

>\begin{align} \frac{∂C}{∂w_{i,j}^\ell}=\frac{∂C}{∂z_j^{\ell+1}} \cdot \frac{∂z_j^{\ell+1}}{∂w_{i,j}^\ell}=\delta_j^{\ell+1} \cdot a_i^{\ell}
\end{align}
\begin{align} \nabla_w^\ell C= 
\begin{pmatrix} \frac{∂C}{\partial w_{1, 1}^\ell} & \frac{∂C}{\partial w_{1,2}^\ell} & ... & \frac{∂C}{\partial w_{1,k}^\ell} \\ 
\frac{∂C}{\partial w_{2,1}^\ell} & \frac{∂C}{\partial w_{2,2}^\ell} & ... & \frac{∂C}{\partial w_{2,k}^\ell} \\
... & ... & ... & ... \\
\frac{∂C}{\partial w_{n,1}^\ell} & \frac{∂C}{\partial w_{n,2}\ell} & ... & \frac{∂C}{\partial w_{n,k}\ell}
\end{pmatrix} =
\begin{pmatrix} \frac{∂C}{\partial z_1^{\ell+1}} ⋅ \frac{∂z_1^{\ell+1}}{\partial w_{1, 1}^\ell} & \frac{∂C}{\partial z_2^{\ell+1}} ⋅ \frac{∂z_2^{\ell+1}}{\partial w_{1, 2}^\ell} & ... & \frac{∂C}{\partial z_k^{\ell+1}} ⋅ \frac{∂z_k^{\ell+1}}{\partial w_{1, k}^\ell} \\ 
\frac{∂C}{\partial z_1^{\ell+1}} ⋅ \frac{∂z_1^{\ell+1}}{\partial w_{2, 1}^\ell} & \frac{∂C}{\partial z_2^{\ell+1}} ⋅ \frac{∂z_2^{\ell+1}}{\partial w_{2, 2}^\ell} & ... & \frac{∂C}{\partial a_k^{\ell+1}} ⋅ \frac{∂z_k^{\ell+1}}{\partial w_{2, k}^\ell} \\
... & ... & ... & ... \\
\frac{∂C}{\partial z_1^{\ell+1}} ⋅ \frac{∂z_1^{\ell+1}}{\partial w_{n, 1}^\ell} & \frac{∂C}{\partial z_2^{\ell+1}} ⋅ \frac{∂z_2^{\ell+1}}{\partial w_{n, 2}^\ell} & ... & \frac{∂C}{\partial z_k^{\ell+1}} ⋅ \frac{∂z_k^{\ell+1}}{\partial w_{n, k}^\ell}
\end{pmatrix}=
\begin{pmatrix}
\nabla_z^{\ell+1}C \cdot a_1^\ell \\
\nabla_z^{\ell+1}C \cdot a_2^\ell \\
... \\
\nabla_z^{\ell+1}C \cdot a_n^\ell
\end{pmatrix}=
\begin{pmatrix}
\delta^{\ell+1} \cdot a_1^\ell \\
\delta^{\ell+1} \cdot a_2^\ell \\
... \\
\delta^{\ell+1} \cdot a_n^\ell
\end{pmatrix}=
\begin{pmatrix}
\delta^{\ell+1}\\
\delta^{\ell+1}\\
... \\
\delta^{\ell + 1}
\end{pmatrix} \circ (A_\ell)^T
\end{align}

>\begin{align} \nabla_b^\ell C= 
\begin{pmatrix} \frac{∂C}{\partial b_{1}^\ell} & \frac{∂C}{\partial b_{2}^\ell} & ... & \frac{∂C}{\partial b_{k}^\ell}
\end{pmatrix} =
\begin{pmatrix} \frac{∂C}{\partial z_1^{\ell}} ⋅ \frac{∂z_1^{\ell}}{\partial b_{1}^\ell} & \frac{∂C}{\partial z_2^{\ell}} ⋅ \frac{∂z_2^{\ell}}{\partial b_{2}^\ell} & ... & \frac{∂C}{\partial z_k^{\ell}} ⋅ \frac{∂z_k^{\ell}}{\partial b_{k}^\ell}
\end{pmatrix}=\delta^\ell
\end{align}

>Dernière équation:
\begin{align}
\nabla_a^\ell C = δ^{\ell+1} \cdot (W_\ell)^T  
\end{align}
Verifier avec theoreme d'inversion locale ou globale


>Descente de Gradient: \begin{align} W_\ell^{(n+1)} = W_\ell^{(n)} - α∇_w^{\ell}C
\end{align}
\begin{align} B_\ell^{(n+1)} = B_\ell^{(n)} - α∇_b^{\ell}C
\end{align}


##Récaptitulatif des formules de rétropropagation:

\begin{align}
δ^\ell=∇_a^lC \circ \sigma'(Z_\ell) && (*)\\
\nabla_a^\ell C = δ^{\ell+1} \cdot (W_\ell)^T && (**) \\
Donnent \ une \ relation \ de \ récurence \ descendante \ sur \ \delta:\\
δ^\ell = (δ^{\ell+1} \cdot (W_\ell)^T) \circ \sigma'(Z_\ell) && (1) \\
\nabla_w^\ell C= \begin{pmatrix}
\delta^{\ell+1}\\
\delta^{\ell+1}\\
... \\
\delta^{\ell + 1}
\end{pmatrix} \circ (A_\ell)^T && (2) \\
\nabla_b^\ell C = \delta^\ell && (3)
\end{align}

##Exemple algorithme du gradient

```python
from matplotlib import pyplot as plt
import numpy as np
from tqdm import tqdm_notebook
```

```python
carre = lambda x: x**2
dcarre = lambda x: 2*x

x0 = 5

def desc_grad(f, df, x0, pas,  tolerance, g, d):
  X = np.linspace(g, d, 1000)
  Y = f(X)
  plt.plot(X,Y)
  i = 0
  while abs(df(x0)) > tolerance:
    x0 = x0 -  pas * df(x0)
    plt.scatter([x0], [f(x0)], c='r', s=4)
    i += 1
  
  print(f"L'algorithme a fait {i} itérations")

desc_grad(carre, dcarre, x0, 0.1, 0.1, -x0, x0)
plt.show()
  
```

```python
x0 = 1

FonctionMoinsTriviale = lambda x: (np.exp(x) + np.cos(x)) * (3-x)
dFonctionMoinsTriviale = lambda x: (np.exp(x) - np.sin(x))*(3-x) - (np.exp(x)+np.cos(x))

desc_grad(FonctionMoinsTriviale, dFonctionMoinsTriviale, x0, 0.1, 0.1, -5, 1)
```

##Création réseau

```python
class Neuralnet:
  def __init__(self, shape, listeActiv):
    actfunc = {
        "relu": (lambda x: np.maximum(0, x)),
        "sigmoid": (lambda x: 1/(1+np.exp(-x)))
    }
    actder = {
        "relu": (lambda x: (x > 0)),
        "sigmoid": (lambda x: np.exp(-x)/(1+np.exp(-x))**2)
    }
    self.actfunc = [actfunc[f] for f in listeActiv]
    self.actder = [actder[f] for f in listeActiv]
    self.prof = len(shape)
    self.m = shape[-1]

    if len(shape) < 2:
      raise Exception("Pas assez de couches")

    #Initialisation Poids
    self.W = []
    for l in range(len(shape)-1):
      self.W.append(np.random.rand(shape[l], shape[l+1])-0.5)

    #Initialisation Biais
    self.B = [None]
    for l in range(1, len(shape)):
      self.B.append(np.random.rand(shape[l])-0.5)



  def evaluer(self, X, get_az = False):
    A = [X]
    Z = [None]
    for l in range(len(self.W)):
      Z.append(np.dot(A[l], self.W[l]) + self.B[l+1])
      A.append(self.actfunc[l](Z[-1]))
    if get_az:
      return A[-1], A, Z
    return A[-1]


  def fit(self, Xset, Yset, epochs, step, batch_s = 1, CY = False):
    nb_batch = len(Xset)//batch_s+1
    if CY:
      Clist = np.zeros((epochs*nb_batch))
      

    for epoch in range(epochs):
      for batchno in tqdm_notebook(range(len(Xset)//batch_s+1), desc=f'Epoch {epoch}'):
        delta_W = [np.zeros_like(self.W[l]) for l in range(len(self.W))]
        delta_B = [np.zeros_like(self.B[l]) for l in range(len(self.B))]
        C = 0

        for batchi in range(batch_s):
          seti = (batchno*batch_s+batchi)%len(Xset)
          X, Yref = Xset[seti], Yset[seti]
          Ycalc, A, Z = self.evaluer(X, True)

          delta = [np.zeros_like(Z[i]) for i in range(len(Z))]
          
          C += np.sum(0.5 * (Ycalc - Yref)**2)/self.m

          delta[self.prof-1] = np.multiply((Ycalc - Yref)/self.m,self.actder[self.prof-2](Z[-1]))
          
          for l in range(self.prof-2, -1, -1):
            gradWlC = np.multiply(np.tile(delta[l+1], (len(A[l]), 1)), np.array([A[l]]).T)
            delta_W[l] += step * gradWlC
            delta_B[l+1] += step * delta[l+1]
            if l != 0:
              delta[l] = np.multiply(np.dot(delta[l+1], self.W[l].T), self.actder[l-1](Z[l]))
        if CY:
            Clist[epoch*nb_batch+batchno] = C/batch_s
        

        for l in range(len(self.W)):
          self.W[l] -= delta_W[l]/batch_s
          self.B[l+1] -= delta_B[l+1]/batch_s
        
      print(C/batch_s)
        
    if CY:
      return Clist
```

```python
Xf = np.linspace(-3,3, 1000)
A = (lambda x: np.maximum(0, x))(Xf)
B = (lambda x: 1/(1+np.exp(-x)))(Xf)
C = (lambda x: (x > 0))(Xf)
D = (lambda x: np.exp(-x)/(1+np.exp(-x))**2)(Xf)
plt.plot(Xf, A)
plt.plot(Xf, B)
plt.plot(Xf, C)
plt.plot(Xf, D)
```

##Partie 2: Classer tous les chiffres

```python
from sklearn.datasets import fetch_openml
import numpy as np

mnist = fetch_openml('mnist_784', version=1)
mnist.keys()

Xdata, Ydata = np.array(mnist["data"])/255, np.array(mnist["target"].astype(np.uint64))
```

```python
Y = np.zeros((Xdata.shape[0], 10))
Y[np.arange(Xdata.shape[0]),Ydata] = 1
Y
```

```python
mnistClassifier = Neuralnet([784, 128, 64, 10], ["relu", "relu","sigmoid"])
```

```python
C = mnistClassifier.fit(Xdata, Y, 3, 0.01, 60, True)
```

```python
plt.scatter(np.arange(0, len(C)), C, s=0.1)
```

```python
fig=plt.figure(figsize=(28,28))
n=300
j=1
for i in Xdata[:n]:
  fig.add_subplot(10,30,j)
  plt.axis('off')
  plt.imshow(i.reshape(28,28), cmap='binary')
  pred = mnistClassifier.evaluer(i)
  plt.title(np.argmax(pred))
  j+=1
plt.show()
```
